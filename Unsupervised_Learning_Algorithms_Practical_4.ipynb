**K-Means Algorithm**

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples = 400, centers = 4, cluster_std = 0.60, random_state = 0)
```

```python
plt.scatter(X[: ,0], X[:, 1], s = 20)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
y_means = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y_means, s = 20, cmap = 'summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha = 0.9)
plt.show()
```

**Agglomerative Clustering**

```python
import pandas as pd
data = pd.read_csv("/content/diabetes.csv")
data.head()
```

```python
#normalize data
from sklearn.preprocessing import normalize
norm_data = normalize(data)
norm_data = pd.DataFrame(norm_data, columns = data.columns)
norm_data.head()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title("Dendrogram of Data")
dendogram = shc.dendrogram(shc.linkage(norm_data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage='ward')
cluster.fit_predict(norm_data)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(norm_data['Glucose'], norm_data['BloodPressure'])
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage='ward')
cluster.fit_predict(norm_data)
```
